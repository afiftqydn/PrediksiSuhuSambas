```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Filter data2 masing-masing tahun
data2_2023 = data2[data2['Tanggal'].dt.year == 2023]
data2_2024 = data2[data2['Tanggal'].dt.year == 2024]

# Titik tertinggi setiap tahun
max_2023 = data2_2023.loc[data2_2023['RR'].idxmax()]
max_2024 = data2_2024.loc[data2_2024['RR'].idxmax()]

# Buat subplot atas-bawah (2 baris, 1 kolom)
fig = make_subplots(
    rows=2, cols=1,
    subplot_titles=("Curah Hujan Tahun 2023", "Curah Hujan Tahun 2024"),
    shared_xaxes=False
)

# Garis tren 2023
fig.add_trace(go.Scatter(
    x=data2_2023['Tanggal'],
    y=data2_2023['RR'],
    mode='lines',
    name='Curah Hujan 2023',
    line=dict(color='blue')
), row=1, col=1)

# Titik tertinggi 2023
fig.add_trace(go.Scatter(
    x=[max_2023['Tanggal']],
    y=[max_2023['RR']],
    mode='markers+text',
    name='Tertinggi 2023',
    marker=dict(color='blue', size=10),
    text=['Tertinggi 2023'],
    textposition='middle right'
), row=1, col=1)

# Garis tren 2024
fig.add_trace(go.Scatter(
    x=data2_2024['Tanggal'],
    y=data2_2024['RR'],
    mode='lines',
    name='Curah Hujan 2024',
    line=dict(color='green')
), row=2, col=1)

# Titik tertinggi 2024
fig.add_trace(go.Scatter(
    x=[max_2024['Tanggal']],
    y=[max_2024['RR']],
    mode='markers+text',
    name='Tertinggi 2024',
    marker=dict(color='green', size=10),
    text=['Tertinggi 2024'],
    textposition='middle right'
), row=2, col=1)

# Layout
fig.update_layout(
    title='Perbandingan Curah Hujan Tahun 2023 dan 2024',
    height=700,
    width=900,
    showlegend=False,
    yaxis_title='Curah Hujan (mm)',
    template='plotly_white'
)

# Format tanggal dan rotasi
fig.update_xaxes(tickformat='%d-%m-%Y', tickangle=-45)

fig.show()
```
